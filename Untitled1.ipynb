```python
import numpy as np
import pandas as pd

# Assuming you have your data in 'marketing_campaign.csv'
df = pd.read_csv('/content/marketing_campaign.csv')

#Age of customer today
df["Age"] = 2021 - df["Year_Birth"]

#Total spendings on various items
df["Spent"] = df["MntWines"]+ df["MntFruits"]+ df["MntMeatProducts"]+ df["MntFishProducts"]+ df["MntSweetProducts"]+ df["MntGoldProds"]

#Deriving living situation by marital status"Alone"
df["Living_With"]=df["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})

#Feature indicating total children living in the household
df["Children"]=df["Kidhome"]+df["Teenhome"]

#Feature for total members in the householde
df["Family_Size"] = df["Living_With"].replace({"Alone": 1, "Partner":2})+ df["Children"]

#Feature pertaining parenthood
df["Is_Parent"] = np.where(df.Children> 0, 1, 0)

#Segmenting education levels in three groups
df["Education"]=df["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})

#For clarity
df=df.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})

#Dropping some of the redundant features
to_drop = ["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
df = df.drop(to_drop, axis=1)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
df = pd.read_csv('/content/marketing_campaign.csv')
ds = df.copy()
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
ds = ds.drop(cols_del, axis=1)
#Identify and encode categorial columns
categorial_cols = ds.select_dtypes(include=['object']).columns.tolist()
for col in categorial_cols:
    ds[col] = ds[col].astype('category')
    ds[col] = ds[col].cat.codes
#Scaling
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

```python
print("Dataframe to be used for further modelling:")
scaled_ds.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
imputer = SimpleImputer(strategy='mean')
scaled_ds_imputed = pd.DataFrame(imputer.fit_transform(scaled_ds), columns=scaled_ds.columns)
pca = PCA(n_components=3)
pca.fit(scaled_ds_imputed)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds_imputed), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
 # Pass ax to sns.scatterplot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Load the dataset
df = pd.read_csv('/content/marketing_campaign.csv')

# Create the 'Spent' column by summing spending-related columns
df['Spent'] = (df['MntWines'] + df['MntFruits'] +
               df['MntMeatProducts'] + df['MntFishProducts'] +
               df['MntSweetProducts'] + df['MntGoldProds'])

# Ensure 'Clusters' column exists (replace with actual clustering logic)
# Dummy assignment for illustration purposes
df['Clusters'] = df['Income'].apply(lambda x: 0 if x > 55000 else (1 if x > 35000 else (2 if x > 25000 else 3)))

# Define the palette to match the image colors
palette = ["#682F2F", "#B9C0C9", "#9F8A78", "#F3AB60"]

# Generate the scatter plot
fig, ax = plt.subplots()  # Create fig and ax explicitly
pl = sns.scatterplot(data=df, x='Spent', y='Income', hue='Clusters', palette=palette, ax=ax) # Pass ax to sns.scatterplot
pl.set_title("Cluster's Profile Based On Income And Spending")
plt.legend(title="Clusters")
ax.yaxis.set_major_locator(ticker.MultipleLocator(20000)) # Use ax.yaxis instead of ax.y_axis
ax.set_ylim(0, 160000)

plt.show()

```
